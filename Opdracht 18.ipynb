```python
from sklearn import tree
import graphviz

def plot_tree_classification(model, features, class_names):
    # Generate plot data
    dot_data = tree.export_graphviz(model, out_file=None, 
                          feature_names=features,  
                          class_names=class_names,  
                          filled=True, rounded=True,  
                          special_characters=True)  

    # Turn into graph using graphviz
    graph = graphviz.Source(dot_data)  

    # Write out a pdf
    graph.render("decision_tree")

    # Display in the notebook
    return graph 
```

```python
from sklearn.tree import DecisionTreeRegressor

features = ['tempo']
dt_regression = DecisionTreeRegressor(max_depth=3) # increase max_depth to see effect in the plot
dt_regression.fit(lastDecade_train[features], lastDecade_train['year'])
```

```python
from sklearn import tree

def plot_tree_regression(model, features):
    # generate plot data
    graph = tree.plot_tree(model,
                          feature_names=features,
                          filled=True)
    
    # display in the notebook
    return graph
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(25,25))

plot_tree_regression(dt_regression, features)
```

Het gene wat me hier opvalt is de donkeroranje en de witte velden, deze hebben weinig samples vergeleken met de andere en ik heb hier eigenlijk geen duidelijke verklaring voor.

```python
def calculate_rmse(predictions, actuals):
    if(len(predictions) != len(actuals)):
        raise Exception('The amount of predictions did not equal the amount of actuals.')
        
    return (((predictions - actuals) ** 2).sum() / len(actuals)) ** (1/2)
```

```python
predictionsOnTrainSet = dt_regression.predict(lastDecade_train[features])
predictionsOnTestSet = dt_regression.predict(lastDecade_test[features])

rmseTrain = calculate_rmse(predictionsOnTrainSet, lastDecade_train['year'])
rmseTest = calculate_rmse(predictionsOnTestSet, lastDecade_test['year'])

print('RMSE on training set ' + str(rmseTrain))
print('RMSE on test set ' + str(rmseTest))
```

De RMSE is ongeveer 3.4 en dit is aan de lage kant maar zeker niet onverklaarbaar, de standaardafwijking is dus aan de lage kant.
